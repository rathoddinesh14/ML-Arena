```python
import pandas as pd

gdp_df = pd.read_csv('indiaGDP.csv')

# label the columns
gdp_df.columns = ['year', 'gdp']

print(gdp_df.head())

print("Shape of the dataframe: ", gdp_df.shape)
print("Data types of the columns: ", gdp_df.dtypes)

# describe the dataframe
print(gdp_df.describe())
```

# Plots

```python
gdp_df.plot(x='year', y='gdp', kind='line', figsize=(10, 5))
```

```python
from sklearn.model_selection import train_test_split

# split the data into train and test
X_train, X_test, y_train, y_test = train_test_split(gdp_df['year'], gdp_df['gdp'], test_size=0.2, random_state=0)

print("Shape of X_train: ", X_train.shape)
print("Shape of y_train: ", y_train.shape)
print("Shape of X_test: ", X_test.shape)
print("Shape of y_test: ", y_test.shape)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression

# Create linear regression object
model = LinearRegression()

# Train the model using the training sets
model.fit(X_train.values.reshape(-1, 1), y_train)
```

```python
import matplotlib.pyplot as plt
```

```python
# predict the values
y_pred = model.predict(X_test.values.reshape(-1, 1))

# metrics
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean squared error: ", mse)
print("R2 score: ", r2)

# plot the actual and predicted values
plt.scatter(X_test, y_test,  color='gray')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.show()
```

# Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor

# Create decision tree regression object
tree_model = DecisionTreeRegressor()

# Train the model using the training sets
tree_model.fit(X_train.values.reshape(-1, 1), y_train)
```

```python
# predict the values
y_pred = tree_model.predict(X_test.values.reshape(-1, 1))

# metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean squared error: ", mse)
print("R2 score: ", r2)

# plot the actual and predicted values
plt.scatter(X_test, y_test,  color='gray')
plt.scatter(X_test, y_pred, color='red')
plt.show()
```

# Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

# Create random forest regression object
forest_model = RandomForestRegressor()

# Train the model using the training sets
forest_model.fit(X_train.values.reshape(-1, 1), y_train)
```

```python
# predict the values
y_pred = forest_model.predict(X_test.values.reshape(-1, 1))

# metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean squared error: ", mse)
print("R2 score: ", r2)

# plot the actual and predicted values
plt.scatter(X_test, y_test,  color='gray')
plt.scatter(X_test, y_pred, color='red')
plt.show()
```
